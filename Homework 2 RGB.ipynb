```python
fig1, ax = plt.subplots(4,1,figsize=(15,8)) # 3 color channels and 1 image plot



ax[0].hist(pika1_data[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, 0th or first channel = Red channel
ax[1].hist(pika1_data[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax[2].hist(pika1_data[:,:,2].flatten(), bins=256, color='blue') # Blue Channel

ax[0].set_xlabel('R channel bins')
ax[1].set_xlabel('G channel bins')
ax[2].set_xlabel('B channel bins')
ax[0].set_ylabel('# of pixels')
ax[1].set_ylabel('# of pixels')
ax[2].set_ylabel('# of pixels')

ax[3].set_facecolor('gray')
ax[3].imshow(pika1_data)

plt.show()
```
